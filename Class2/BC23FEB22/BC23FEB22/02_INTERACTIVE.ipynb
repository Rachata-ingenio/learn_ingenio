<h4>EX01: Creating a Staic Scatter Plot</h4>

```python
import seaborn as sns
mpg_df = sns.load_dataset("mpg"); mpg_df.head()
```

```python
ax = sns.scatterplot(x = 'weight', y = 'mpg', data = mpg_df)
```

<h4>EX02: Creating a Static Hexagonal Binning Plot</h4>

```python
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
mpg_df = sns.load_dataset("mpg")
sns.set(style = 'ticks')
sns.jointplot(mpg_df['weight'], mpg_df['mpg'], kind = 'hex', color = '#4CB391')
```

<h4>EX03: Creating a Static Contour Plot</h4>

```python
import seaborn as sns
mpg_df = sns.load_dataset("mpg")
sns.set_style('white')
sns.kdeplot(mpg_df['weight'], mpg_df['mpg'], shade = True)
```

<h4>EX04: Creating a Static Line Plot</h4>

<h4>EX05: Presenting Data across Time with Multiple Line Plots</h4>

```python
import seaborn as sns
flights_df = sns.load_dataset("flights"); flights_df.head()
```

```python
for m in flights_df['month'].unique(): 
    sns.lineplot(x = 'year',
                 y = 'passengers', 
                 data = flights_df[flights_df['month'] ==  m])
```

<h4>EX06: Creating and Exploring a Static Heatmap</h4>

```python
import seaborn as sns
flights_df = sns.load_dataset("flights")
```

```python
df_pivoted = flights_df.pivot('month', 'year', 'passengers'); df_pivoted.head()
```

```python
ax0 = sns.heatmap(df_pivoted)
```

```python
ax1 = sns.clustermap(df_pivoted, col_cluster = False, row_cluster = True)
```

<h4>EX07: Creating Linkage in Static Heatmaps</h4>

<ul>
    <li>If we define the distance between two clusters as the distance between the two points across the clusters closest to each other, the rule is called <b>single linkage</b>.
    <li>If the rule is to define the distance between two clusters as the distance between the points farthest from each other, it is called <b>complete linkage</b>.
    <li>If the rule is to define the distance as the average of all possible pairs of rows in the two clusters, it is called <b>average linkage</b>.


```python
import seaborn as sns
flights_df = sns.load_dataset("flights")
```

```python
df_piovted = flights_df.pivot('month', 'year', 'passengers')
```

```python
ax0 = sns.clustermap(df_pivoted, col_cluster = False,
                     metric = 'correlation', method = 'average')
ax1 = sns.clustermap(df_pivoted, col_cluster = False,
                     metric = 'correlation', method = 'complete')
ax2 = sns.clustermap(df_pivoted, col_cluster = False, 
                     metric = 'correlation', method = 'single')
```

<h3>Creating Plots That Present Summary Statistics of Your Data</h3> 

<h4>EX08: Histogram Revisited</h4>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
mpg_df = sns.load_dataset("mpg")
ax0 = sns.displot(mpg_df['weight'], bins = 50)
plt.axvline(x = np.mean(mpg_df['weight']), color = 'b', label = 'mean')
plt.axvline(x = np.median(mpg_df['weight']), color = 'navy', ls = '--', label = 'median')
plt.legend(loc = 'upper right')
```

```python
ax1 = sns.boxplot(x = 'model_year', y = 'mpg', data = mpg_df)
```

```python
mpg_df['model_decade'] = (np.floor(mpg_df['model_year'] / 10) * 10).astype(int)
mpg_df.head()
```

```python
ax2 = sns.boxplot(x = 'model_decade', y = 'mpg', data = mpg_df)
```

```python
ax3 = sns.boxplot(x = 'model_decade', y = 'mpg', hue = 'origin', data = mpg_df)
```

```python
ax4 = sns.violinplot(x = 'model_decade', y = 'mpg', hue = 'origin', data = mpg_df)
```
